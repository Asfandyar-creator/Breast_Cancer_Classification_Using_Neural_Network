# Breast Cancer Classification using Simple Neural Network

Importing the Dependencies

```python
import pandas as pd
import numpy as np
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Pre-Processing

```python
# Loading the dataset from sklearn
breat_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breat_cancer_dataset)
```

```python
# Loading the dataset into Pandas DataFrame
data_frame = pd.DataFrame(breat_cancer_dataset.data, columns=breat_cancer_dataset.feature_names)
```

```python
# Printing first 5 rows of dataset
data_frame.head()
```

```python
# Adding the "target" column to DataFrame
data_frame['label'] = breat_cancer_dataset.target
```

```python
# Printing last 5 rows of DataFrame
data_frame.tail()
```

```python
# Number of rows and columns
data_frame.shape
```

```python
# Getting the information about DataFrame
data_frame.info()
```

```python
# Checking for missing values
data_frame.isna().sum().sort_values()
```

```python
# Statistical measures
data_frame.describe()
```

```python
# Checking the distribution of "target" variable
data_frame['label'].value_counts()
```

1 -> Benign

0 -> Malignant

```python
data_frame.groupby('label').mean()
```

Separating the Features and Target variable

```python
X = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
print(X)
print(y)
```

Splitting the Training and Test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1111)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

**Building the Neural Network**

```python
import tensorflow as tf
tf.random.set_seed(1111)
from tensorflow import keras
```

```python
# Setting up layers of neural network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# Compiling the neural network

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Training the neural network
history = model.fit(X_train_scaled, y_train, validation_split=0.1, epochs=10)
```

Visualizing Accuracy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['Training Data', 'Validation Data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')

plt.legend(['Training Data', 'Validation Data'], loc='upper right')
```

Accuracy of Model on Test Data

```python
loss, accuracy = model.evaluate(X_test_scaled, y_test)
print(accuracy)
```
